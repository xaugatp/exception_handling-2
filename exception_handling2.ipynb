```python
'''Exception class is the base class for all exceptions. When creating a custom 
exception, it is good practice to inhertit from the exception class or a more specificed subclass 
of exception, for several reasons:
Inheritance from Common Base Class:

Inheriting from the Exception class ensures that your custom exception is part of the common hierarchy for all exceptions in Python. This hierarchy makes it easier to catch and handle exceptions in a more generic way, as you can catch multiple exceptions by catching their common ancestor.
Consistent Exception Handling:

By inheriting from the Exception class, your custom exception follows the established conventions for exception handling in Python. This consistency makes it easier for developers to understand and work with your custom exceptions, as they can apply similar handling patterns used for built-in exceptions.
Compatibility with Exception Handling Mechanisms:

Python's exception handling mechanisms, such as try, except, and finally blocks, are designed to work with exceptions derived from the Exception class. When you create a custom exception by inheriting from Exception, you can seamlessly integrate it into these mechanisms.
Customization and Clarity:

By inheriting from Exception, you gain the ability to customize your exception class with additional attributes or methods if needed. This can be useful for providing more information about the exception or implementing specific behaviors.
Here's an example illustrating the creation of a custom exception by inheriting from the Exception class:

python
Copy code
class CustomError(Exception):
    def __init__(self, message="A custom error occurred"):
        self.message = message
        super().__init__(self.message)

# Example usage:
try:
    raise CustomError("This is a custom exception.")
except CustomError as e:
    print(f"Caught a custom exception: {e}")'''
```

```python
'''The LookupError class is a base class for exceptions that occur when a key or index is not found during a lookup operation. It serves as the parent class for more specific lookup-related exceptions in Python. Two common exceptions derived from LookupError are KeyError and IndexError. Let's discuss these errors with examples:

1. KeyError:
Description: Raised when a dictionary key is not found.

Example:

python
Copy code
my_dict = {'name': 'John', 'age': 25}

try:
    value = my_dict['city']  # This will raise KeyError
except KeyError as e:
    print(f"Error: {e}")
In this example, we attempt to access the value associated with the key 'city' in the dictionary my_dict. Since 'city' is not a key in the dictionary, a KeyError is raised. The program catches the exception in the except block, prints an error message, and continues executing.

2. IndexError:
Description: Raised when a sequence (e.g., list, tuple) index is out of range.

Example:

python
Copy code
my_list = [1, 2, 3]

try:
    value = my_list[5]  # This will raise IndexError
except IndexError as e:
    print(f"Error: {e}")'''
```

```python
'''ImportError is a base class for exceptions raised when an import statement fails to import a module or when an attempt is made to use a module or name that is not present. It is a general exception that encompasses various import-related errors. One specific subclass of ImportError introduced in Python 3.6 and later is ModuleNotFoundError.

1. ImportError:
Description: Raised when an import statement cannot locate or load the specified module.

Example:

python
Copy code
try:
    import non_existent_module  # This will raise ImportError
except ImportError as e:
    print(f"Error: {e}")
In this example, the program attempts to import a module named non_existent_module, which doesn't exist. This will result in an ImportError. The program catches the exception in the except block, prints an error message, and continues executing.

2. ModuleNotFoundError:
Description: A subclass of ImportError introduced in Python 3.6 and later. Specifically raised when an import statement cannot find the specified module.

Example:

python
Copy code
try:
    import non_existent_module  # This will raise ModuleNotFoundError
except ModuleNotFoundError as e:
    print(f"Error: {e}")
This example is the same as the previous one, but it explicitly catches ModuleNotFoundError. In modern Python versions (3.6 and later), using ModuleNotFoundError allows for more specific handling of cases where a module is not found during the import process.'''
```
